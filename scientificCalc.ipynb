```python
#This is the main function to be called to run the calculator. Every other function is linked to it.
def calculator():
    calcMemory = []
    calcInstructions()
    allOperators = ('M', '*', 'A', '+', 'D', '/', 'S', '-', 'E', '^', 'SQR', 'FD', 'MO','MOP', 'AVE', 'SUM', 'MED', 'MX', 'MR', 'Q')
    while True:
        operation = input ('what would you like to do? ').strip().upper()
        
        if operation not in allOperators:
            print ('Invalid operator. Please enter valid operator')
            continue
        elif operation == 'MOP':
            multiOperation(calcMemory)
        elif operation == 'Q':
            return None
        elif operation == 'MR':
            result = calculatorMemory(operation, calcMemory)
            print (result)
        else:
            result = calculate(operation, calcMemory)
            print (result) 
    
#This function is to print the calculator instructions to help the user
def calcInstructions():
    calculatorFunctions = ''' INSTRUCTION!!! 
    For the 'what would you like to do?' prompt, please type in CAPS ONLY for letters or use the symbols where applicable.
    1. M or * to multiply only 2 numbers
    2. A or + to add only 2 numbers 
    3. D or / to divide 2 numbers
    4. S or - to subtract one number from another
    5. E or ^ for exponential of a number by another number
    6. SQR for square root of a number
    7. FD for floor division of a number
    8. MO for modulus operation of a number
    9. AVE for mean of a list of numbers
    10. SUM for sum of a list of numbers
    11. MED for median of a list of numbers
    12. MX to multiply a list of numbers
    13. MR to access calculator memory
    14. MOP to carry out multiple operations, e.g 2+3-4*9/10
    15. Q to Quit/Exit calculator
    '''
    print (calculatorFunctions)

#This function is for single operator operations like 2+3, 5/10, 6*7, etc
def singleOperations(operation):
       
    value1 = float(input('type the first number: '))
    value2 = float(input('type the second number: '))

    if operation == 'M' or operation == '*':
        result = value1 * value2
        userInput = (f'{value1} * {value2} = {result}')

    elif operation == 'A' or operation == '+':
        result = value1 + value2
        userInput = (f'{value1} + {value2} = {result}')

    elif operation == 'D' or operation == '/':
        if value2 == 0:
            result = ('undefined')
            userInput = (f'{value1} / {value2} = {result}')
        else:
            result = value1 / value2
            userInput = (f'{value1} / {value2} = {result}')

    elif operation == 'S' or operation == '-':
        result = value1 - value2
        userInput = (f'{value1} - {value2} = {result}')

    elif operation == 'E' or operation == '^':
        result = value1 ** value2
        userInput = (f'{value1} ** {value2} = {result}')

    elif operation == 'FD':
        result = value1 // value2
        userInput = (f'{value1} // {value2} = {result}')

    elif operation == 'MO':
        result = value1 % value2
        userInput = (f'{value1} % {value2} = {result}')
    
    return result, userInput

#This function is for operations that require a list of numbers, separated by a comma                    
def listOperations(operation):
    
    listValues = input('type the list of numbers separated by a comma: ')
    valuesList = listValues.split(',')
    valuesList = list(float(num.strip()) for num in valuesList)
    
    if operation == 'AVE':
        result = sum(valuesList)/ len(valuesList)
        userInput = (f'mean {valuesList} = {result}')

    elif operation == 'SUM':
        result = sum(valuesList)
        userInput = (f'sum {valuesList} = {result}')

    elif operation == 'MX':
        result = 1
        for num in valuesList:
            result = num * result
        userInput = (f'product {valuesList} = {result}')

    elif operation == 'MED':
        result = statistics.median(valuesList)
        userInput = (f'median {valuesList} = {result}')
                
    return result, userInput

#This function houses the calculations from any of the above 2 functions and also stores in the calculator memory
def calculate(operation, calcMemory):
    if operation in ('M', '*', 'A', '+', 'D', '/', 'S', '-', 'E', '^','FD', 'MO'):
        result, userInput = singleOperations(operation)
        
    elif operation in ('AVE', 'SUM', 'MED', 'MX'):
        result, userInput = listOperations(operation)
    
    elif operation == 'SQR':
        value = float(input('type the number: '))
        result = math.sqrt(value)
        userInput = (f'SQR {value} = {result}')
        
    calcMemory.append(userInput)
    #calcMemory = calcMemory[-10:10]
    if len(calcMemory) == 11:
        del calcMemory[0]
    else:
        calcMemory = calcMemory
        
    return result

#This function only runs if the user wants to perform multiple operations without breaking the operation. 
#The result is also stored in the calculator memory after the operation is done
def multiOperation(calcMemory):
    multiOpGuide = '''for the 'operator' prompt, enter '*' to multiply, '+' for addition, '-' for subtraction, 
    '/' for division, and '^' for exponential'''
    value1 = float(input('num1 ')) 
    opr = input('operator ')
    value2 = float(input('num2 '))
    
    if opr == '+':
        result = value1 + value2
    elif opr == '-':
        result = value1 - value2
    elif opr == '*':
        result = value1 * value2
    elif opr == '/':
        if value2 == 0:
            print('undefined')
            multiOperation(calcMemory)
        else:
            result = value1 / value2
    elif opr == '^':
        result = value1 ** value2
    print(result)
    while True:
        result = result
        opr = input('operator ')
        if opr == '=':
            print(result)
            userInput = result
            calcMemory.append(userInput)
            if len(calcMemory) == 11:
                del calcMemory[0]
            else:
                calcMemory = calcMemory
         
            return None
        valueX = float(input('numX '))
        
        if opr == '+':
            result = result + valueX
        elif opr == '-':
            result = result - valueX
        elif opr == '*':
            result = result * valueX
        elif opr == '/':
            if valueX == 0:
                result = 'undefined'
                continue
            else:
                result = result / valueX
        elif opr == '^':
            result = result ** valueX
        print(result)
        
#This function helps to recall any of the last 10 operations which is housed in the calculator memory   
def calculatorMemory(operation, calcMemory):
    memoryGuide = '''For the 'which of the last 10 operations do you want to see?' prompt, 
        1. type 1 to see the last operation
        2. type 2 to see the last two operations
        3. type 3 to see the last three operations
        ...
        4. type n to see the last nth operation, up to the last 10 operations'''
        
    print(memoryGuide)
    
    recall = int(input('which of the last 10 operations do you want to see? '))
    
    if len(calcMemory) == 0:
        userInput = 'Calculator memory is empty'
    else:
        idx = recall* -1
        if len(calcMemory) < recall:
            print ('This memory is not available')
            return None
        else:
            userInput = calcMemory[idx]
            
    return userInput
```
